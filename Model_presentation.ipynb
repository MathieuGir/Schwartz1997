The PDE for futures prices gives the following equation:

$$\ln F(S_0,\delta, r, T) = \ln S_0 - r\,A(T) - \delta\,B(T) + C(T)$$
$$ \frac{dS_t}{S_t} = (r - \delta) dt + \sigma_1 dz_1 $$
$$ d \delta = \kappa (\hat{\alpha} - \delta) dt + \sigma_2 dz_2 $$
$$ d r_t = a (m^* - r_t) + \sigma_3 dz_3 $$

with $dz_1 dz_2 = \rho_1 $, $dz_2 dz_3 = \rho_2 $ and $dz_1 dz_3 = \rho_3 $

The PDE for futures prices gives the following equation: 
$$ ln F(S, \delta, r,T ) = ln(S) - r A(t) - \delta B(T) + C(T) $$

With:
\begin{align}
A(T) &= \frac{1 - e^{-aT}}{a}, \\
B(T) &= \frac{1 - e^{-\kappa T}}{\kappa}, \\
C(T) &= \frac{\kappa \hat{\alpha}\,\big((1 - e^{-\kappa T}) - \kappa T\big)}{\kappa^{2}}
      - \frac{\sigma_2^{2}\,\big(4(1 - e^{-\kappa T}) - (1 - e^{-2\kappa T}) - 2\kappa T\big)}{4\kappa^{3}} \\
&\quad - \frac{a m^{*}\,\big((1 - e^{-aT}) - aT\big)}{a^{2}}
      - \frac{\sigma_3^{2}\,\big(4(1 - e^{-aT}) - (1 - e^{-2aT}) - 2aT\big)}{4a^{3}} \\
&\quad + \frac{\kappa^{2}(1 - e^{-aT}) + a^{2}(1 - e^{-\kappa T}) - \kappa a^{2}T - a\kappa^{2}T}
              {\kappa^{2} a^{2}(\kappa + a)} .
\end{align}

Assuming $\rho_2 = \rho_3 = 0$ (i.e. independence of the short rate process compared to commodity price and convenience yield)

# Measurement equation and Kalman Filter

\begin{align}
\begin{pmatrix}
\ln F(T_1) \\
\ln F(T_2) \\
\vdots      \\
\ln F(T_n)
\end{pmatrix}
&=
\begin{pmatrix}
1 & -A(T_1) & B(T_1) \\
1 & -A(T_2) & B(T_2) \\
\vdots & \vdots & \vdots \\
1 & -A(T_n) & B(T_n)
\end{pmatrix}
\begin{pmatrix}
X_t \\
\delta_t \\
r_t
\end{pmatrix}
+
\begin{pmatrix}
C(T_1) \\
C(T_2) \\
\vdots \\
C(T_n)
\end{pmatrix}
+
\begin{pmatrix}
\varepsilon_1 \\
\varepsilon_2 \\
\vdots \\
\varepsilon_n
\end{pmatrix}.
\end{align}


Which is re-written:
$$ y_t = Z_t x_t + dt + \varepsilon_t $$

The transition equation for the latent state variables is the following:

\begin{align}
\begin{pmatrix}
X_t \\
\delta_t \\
r_t
\end{pmatrix}
&=
\begin{pmatrix}
1 & -\Delta t & \Delta t \\
0 & 1 - \kappa \Delta t & 0 \\
0 & 0 & 1 - a \Delta t
\end{pmatrix}
\begin{pmatrix}
X_{t-1} \\
\delta_{t-1} \\
r_{t-1}
\end{pmatrix}
+
\begin{pmatrix}
-\frac{\sigma_1^2}{2}\,\Delta t \\
\kappa \hat{\alpha}\,\Delta t \\
a m^*\,\Delta t
\end{pmatrix}
+
\begin{pmatrix}
\eta_1 \\
\eta_2 \\
\eta_3
\end{pmatrix}.
\end{align}


$\eta_t \begin{pmatrix}
\eta_1 \\
\eta_2 \\
\eta_3
\end{pmatrix},
\qquad
\eta_t \sim \mathcal{N}(0,\, Q),$ where $ Q = \begin{pmatrix}
\sigma_1^2 \Delta t & \rho_1 \sigma_1 \sigma_2 \Delta t & 0 \\
\rho_1 \sigma_1 \sigma_2 \Delta t & \sigma_2^2 \Delta t & 0 \\
0 & 0 & \sigma_3^2 \Delta t
\end{pmatrix} $
